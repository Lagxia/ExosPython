```python
# ON SOUHAITE MODIFIER L'ECHELLE DE MESURE DES DIFFERENTS ACIDES
# En divisant les valeurs des colonnes concernées par 100
Liste_des_Acides =['Palmitique', 'Palmitoleique', 'Stearique', '0leique',
                   'Linoleique', 'Linolenique', 'Arachidique', 'Eicosenoide']
Huile_Olive_100 = Huile_Olive_df[Liste_des_Acides].apply(lambda x: x/100.0)
Huile_Olive_100.head(5)
```

```python
# ON PLACE ENSUITE CETTE LISTE DE VALEURS DANS LE DATAFRAME 
Huile_Olive_df[Liste_des_Acides] = Huile_Olive_100
Huile_Olive_df.head(5)
```

```python
# STATISTIQUES DES QUANTITES D'ACIDES PAR REGION
Groupe_Par_Region = Huile_Olive_df.groupby('Nom_Region')
Groupe_Par_Region_Stats = Groupe_Par_Region.describe()
Groupe_Par_Region_Stats.head(20)
```

```python
# STATISTIQUES SUR LES ECARTS TYPES DES QUANTITES D'ACIDES PAR REGION
HuileStd = Huile_Olive_df.groupby('Nom_Region').std()
HuileStd
```

```python
# STATISTIQUES SUR LES MOYENNES DES QUANTITES D'ACIDES PAR REGION
import numpy as np
HuileMean = Groupe_Par_Region.aggregate(np.mean)
HuileMean
```

```python
# STATISTIQUES SUR LES ECARTS TYPES DES QUANTITES D'ACIDES PAR REGION (en renommant les colonnes)
renamedict_std = {k:k+"_EcTyp" for k in Liste_des_Acides}
HuileStd.rename(columns = renamedict_std, inplace = True)
HuileStd
```

```python
# STATISTIQUES SUR LES MOYENNES DES QUANTITES D'ACIDES PAR REGION (en renommant les colonnes)
renamedict_mean = {k:k+"_Moy" for k in Liste_des_Acides}
HuileMean.rename(columns = renamedict_mean, inplace = True) 
HuileMean
```

```python
# JOINTURE DE CES DEUX TABLEAUX POUR STATISTIQUES ACIDES PALMITIQUES
HuilePalmitiqueMean = HuileMean[['Palmitique_Moy']] 
HuilePalmitiqueStd = HuileStd[['Palmitique_EcTyp']] 
PalmitiqueParRegion = HuilePalmitiqueMean.join(HuilePalmitiqueStd)
PalmitiqueParRegion
```
