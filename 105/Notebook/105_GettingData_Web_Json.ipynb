```python
import requests
from bs4 import BeautifulSoup
```

```python
# Beautiful Soup
html = requests.get('http://www.aston-ecole.com/').text
soup = BeautifulSoup(html)
print(soup)
```

```python
first_paragraph = soup.find('title')
first_paragraph
```

```python
first_paragraph_text = soup.find('title').text
first_paragraph_text
```

```python
first_paragraph_words = soup.find('title').text.split()
first_paragraph_words
```

```python
body_text = soup.find('body').text
body_text
```
