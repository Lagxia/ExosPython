```python
# Gestion des imports
import configparser
import pandas as pd
import xlsxwriter
```

```python
# Gestion de la configuration
config = configparser.ConfigParser()
config.read('../Configuration.conf')
```

```python
# Obtention du chemin de travail
cheminressources = config.get('Chemin', 'resources')
print("Le chemin des fichiers de ressources est : \n", cheminressources)
```

```python
# CHARGEMENT FICHIER Dâ€™ENTREE
Huile_Olive_df = pd.read_csv(cheminressources + 'Huile_Olive.csv', sep = ',')
Huile_Olive_df.head()
```

```python
# TRUC POUR RENOMMER LES COLONNES (quand il y en a beaucoup) : EN FRANCAIS
tabnomcolonnes = ['Nom_Region', 'Num_Region', 'Num_Endroit', 'Palmitique',
                  'Palmitoleique', 'Stearique', '0leique', 'Linoleique',
                  'Linolenique', 'Arachidique', 'Eicosenoide']
Huile_Olive_df.rename(columns = dict(zip(Huile_Olive_df.columns, tabnomcolonnes)),inplace = True)
Huile_Olive_df.head()
```

```python
# CHARGEMENT DE FICHIER Dâ€™ENTREE
Ventes_Voitures_df = pd.read_csv(cheminressources + 'Voitures_Vente.csv', sep = ',', parse_dates = ['date'])
Ventes_Voitures_df.head()
```

```python
# CHARGEMENT DE FICHIERS Dâ€™ENTREE
Presidentielle2017Tour1_df = pd.read_csv(cheminressources + 'Presidentielle_2017_PremierTour.csv', sep = ';')
Presidentielle2017Tour1_df.head()
```

```python
# CHARGEMENT ENTIER D'UN FICHIER JSON (En ligne) DANS UN dataframe
with open(cheminressources + 'Recettes.json', 'r') as file:
    # Extraction de chaque ligne
    data = (line.strip() for line in file)
    # Reformatage de chaque ligne en Ã©lÃ©ment de liste JSON
    data_json = "[{0}]".format(','.join(data))
# Lecture PANDAS comme un fichier JSON
recettes = pd.read_json(data_json)
recettes.head()
```

```python
# CHARGEMENT DU FICHIER Datanova de LA POSTE (Code commune INSEE - Code postal - GÃ©opoints)
laposte_df = pd.read_csv('http://datanova.legroupe.laposte.fr/explore/dataset/laposte_hexasmal/download/?format=csv&timezone=Europe/Berlin&use_labels_for_header=true', sep = ';')
laposte_df.head()
```

# Ecritures

```python
# Obtention du chemin de travail
cheminsorties = config.get('Chemin', 'output')
print("Le chemin des fichiers de sorties est : \n", cheminsorties)
```

```python
# ECRITURE DU FICHIER DE SORTIE (FORMAT CSV)
laposte_df.to_csv(cheminsorties + 'Base_LaPoste_Brut.csv')
laposte_df.head()
```

```python
# ECRITURE DU FICHIER DE SORTIE (FORMAT CSV)
laposte_df.to_csv(cheminsorties + 'Base_LaPoste.csv',
                  index = None,
                  sep = ';',
                  na_rep = 'NaN',
                  float_format = 'string',
                  header = True,
                  encoding = 'UTF-8')
```

```python
# ECRITURE DU FICHIER DE SORTIE (FORMAT CSV)
Ventes_Voitures_df.to_csv(cheminsorties + 'Ventes_Voitures_Brut.csv')
Ventes_Voitures_df.head()
```

```python
# ECRITURE DU FICHIER DE SORTIE (FORMAT CSV)
Ventes_Voitures_df.to_csv(cheminsorties + 'Ventes_Voitures.csv',
                          index = True,
                          index_label = None,
                          header = False,
                          encoding = 'UTF-8',
                          date_format = '%d/%m/%Y',
                          decimal = ',')
```

```python
# ECRITURE DU FICHIER DE SORTIE (EN EXCEL)
writer = pd.ExcelWriter(cheminsorties + 'Ventes_Voitures.xlsx')
Ventes_Voitures_df.to_excel(writer, 'Voitures')
writer.save()
```
