```python
# AFFICHAGE DES INFORMATIONS CONCERNANT LE NOMBRE DE VOIX PAR DEPARTEMENT
Nb_Voix = Fichier_OK_df[['Voix','Libellé du département']]
Groupage_Nb_Voix = Nb_Voix.groupby('Libellé du département').sum()
Groupage_Nb_Voix.head()
```

```python
# VISUALISATION : HISTOGRAMME SIMPLE DU TOTAL DE VOIX PAR DEPARTEMENT
%matplotlib inline
import matplotlib.pyplot as plt
Ma_Visu = Groupage_Nb_Voix.plot(kind='bar',  figsize=(30, 5))
```

```python
# AFFICHAGE DES INFORMATIONS CONCERNANT LE NOMBRE DE VOIX PAR DEPARTEMENT PAR CANDIDAT
Nb_Voix_Candidat = Fichier_OK_df[['Voix','Libellé du département','Nom']]
Nb_Voix_Candidat.head()

# AFFICHAGE DU TOTAL DE VOIX PAR DEPARTEMENT PAR CANDIDAT (PAR LIGNE)
Groupage_Nb_Voix_Candidat = Nb_Voix_Candidat.groupby(['Libellé du département','Nom']).sum()
Groupage_Nb_Voix_Candidat.head(10)
```

```python
# AFFICHAGE DU TOTAL DE VOIX PAR DEPARTEMENT PAR CANDIDAT (SOUR FORME DE TABLEAU)
Groupage_Nb_Voix_Candidat.unstack().head()
```

```python
# VISUALISATION : HISTOGRAMME DU TOTAL DE VOIX PAR DEPARTEMENT PAR CANDIDAT (AVEC LEGENDES)
Ma_Visu = Groupage_Nb_Voix_Candidat.unstack().plot(kind='bar', stacked = True,   figsize=(30, 15), title = "Total des voix par candidat")
Ma_Visu.set_xlabel("Departement")
Ma_Visu.set_ylabel("Nb Voix")
```
