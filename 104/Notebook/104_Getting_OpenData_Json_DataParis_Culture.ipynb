```python
# IMPORTS
import pandas as pd
from pandas import Series,DataFrame
from pandas.io.json import json_normalize
import numpy as np
import json
import urllib
```

```python
##################################### Base de données : CULTURE #######################################
################## Musées de France à Paris
url = "https://data.opendatasoft.com/api/records/1.0/search/?dataset=liste-musees-de-france-a-paris%40parisdata&facet=cp&rows=100"
```

```python
start = 0
offset = 100

response = urllib.request.urlopen(url)
data = response.read()
json_data = json.loads(data)
values = json_data['records']

nb_values = json_data['nhits']

print (str(len(values)) + "/" + str(nb_values))

while (len(values) < nb_values):
    start += offset
    next_url = url + "&start=" + str(start)
    response = urllib.request.urlopen(next_url)
    data = response.read()
    json_data = json.loads(data)
    values += json_data['records']
    print (str(len(values)) + "/" + str(nb_values))

data_df = json_normalize(values)
```
