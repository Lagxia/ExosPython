```python
# 'r' means read-only
with open('../resources/Fables_La_Fontaine.txt','r') as file:
    compteurnomsfables = 0
    for line in file:
        if len(line) > 1 :
            print(line)
            compteurnomsfables += 1
    print("\nTotal de fables de La Fontaine : ", compteurnomsfables)
```

```python
with open('../resources/Fables_La_Fontaine.txt','r') as file:
    word_counts = {}
    for line in file:
        if len(line) > 1 :       
            for word in line.strip().split():
                if word in word_counts:
                    word_counts[word] += 1
                else:
                    word_counts[word] = 1
    print(word_counts)
```

```python
from collections import Counter

with open('../resources/Fables_La_Fontaine.txt','r') as file:
    mots = []
    for line in file:
        for word in line.strip().split():
            mots.append(word)

word_counts = Counter(mots)
for word, count in word_counts.most_common(20):
    print (word, count)
```
