```python
# FONCTION : TRANSFORMATION DE L'OBJET EN FLOAT (REMLACEMENT DE TOUS LES ',' par '.' PUIS TRANSFO EN FLOAT)
def transfoobjet(monobjet):
    monobjet = str(monobjet).replace(',','.')
    monobjet = float(monobjet)
    return monobjet
```

```python
# ETUDE ET VISUALISATION DES VOTES DANS UN DEPARTEMENT
Fichier_Dept_df = Fichier_OK_df[Fichier_OK_df['Code du département'] == '59']
Fichier_Dept_df['% Voix/Ins'] = Fichier_Dept_df['% Voix/Ins'].apply(transfoobjet)
Fichier_Dept_df.describe()
```

```python
# AFFICHAGE ET MOYENNES DES % de Voix par inscrits PAR CANDIDAT PAR CANTON (VISUALISATIONS AVEC C3JS)
SelectionC3JS = Fichier_Dept_df[['Code du canton', 'Nom', '% Voix/Ins']].groupby(['Code du canton', 'Nom']).sum().unstack()
SelectionC3JS
```

```python
# VISUALISATIONS AVEC C3js : AFFICHAGE
MonAbscisse = list(SelectionC3JS.index.values.astype('str'))
MonLabel1 = 'MÉLENCHON'
Label1 = list(SelectionC3JS['% Voix/Ins', MonLabel1].values)
MonLabel2 = 'MACRON'
Label2 = list(SelectionC3JS['% Voix/Ins', MonLabel2].values)
MonLabel3 = 'FILLON'
Label3 = list(SelectionC3JS['% Voix/Ins', MonLabel3].values)
MonLabel4 = 'LE PEN'
Label4 = list(SelectionC3JS['% Voix/Ins', MonLabel4].values)

chart = c3.BarChart()
chart.plot(Label1, label = MonLabel1)
chart.plot(Label2, label = MonLabel2)
chart.plot(Label3, label = MonLabel3)
chart.plot(Label4, label = MonLabel4)
chart.set_xticklabels(MonAbscisse)
chart.ylabel('% Voix par inscrit')
chart.show()
```
