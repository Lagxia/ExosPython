## Exercice "joueurs de football"

```python
df = pd.read_csv('https://raw.githubusercontent.com/rasbt/python_reference/master/Data/some_soccer_data.csv')
df
```

```python
# Format du dataframe
print("Le format du dataframe est : ", df.shape)
print("Le dataframe contient {} lignes.".format(df.shape[0]))
print("Le dataframe contient {} colonnes.".format(df.shape[1]))
```

```python
# Convertir les noms de colonnes en minuscule
df.columns = [c.lower() for c in df.columns]
df.head(3)
```

```python
# Convertir les noms de colonnes en minuscule (autre méthode à creuser)
df.rename(columns = lambda x : x.lower())
df.tail(3)
```

```python
# Renommer les colonnes en français
# 'player' : 'joueur', 'salary' : 'salaire', 'gp': 'matchs_joues', 'g': 'buts_marqués', 'a': 'soutiens',
# 'sot': 'tirs_cadres', 'ppg': 'points_par_match', 'p': 'points'
df = df.rename(columns = {'p': 'points', 
                          'gp': 'matchs_joues',
                          'sot': 'tirs_cadres',
                          'g': 'buts_marqués',
                          'ppg': 'points_par_match',
                          'a': 'soutiens',
                          'player' : 'joueur',
                          'salary' : 'salaire'})

df.tail(3)
```

```python
# Transformer le salaire de $xx.xm à xx.x
df['salaire'] = df['salaire'].apply(lambda x: x.strip('$m'))
df.tail()
```

```python
# Pour préparer la scission de la colonne 'joueur' en 3 colonnes 'joueur' 'poste' et 'equipe'
# créer les deux nouvelles colonnes
df['equipe'] = pd.Series('', index = df.index)
df.insert(loc = 8, column = 'poste', value='') 
df.tail(3)
```

```python
# Traitement de la colonne 'joueur'

def traitement_colonne_joueur(text):
    nom, reste = text.split('\n')
    poste, equipe = [x.strip() for x in reste.split(' — ')]
    return pd.Series([nom, equipe, poste])

df[['joueur', 'equipe', 'poste']] = df.joueur.apply(traitement_colonne_joueur)

# Autre approche (mais évitons les boucles sur les gros dataframes):
#
#for idx,row in df.iterrows():
#    nom, poste, equipe = traitement_colonne_joueur(row['joueur'])
#    df.ix[idx, 'joueur'], df.ix[idx, 'poste'], df.ix[idx, 'equipe'] = nom, poste, equipe
    
df.tail(3)
```

```python
# Convertir les infos des colonnes de texte en minuscule
colonnes = ['joueur', 'poste', 'equipe']
df[colonnes] = df[colonnes].applymap(lambda x: x.lower())
df.head()
```

```python
# Format du dataframe
print("Le format du dataframe est maintenant : ", df.shape)
print("Le dataframe contient maintenant {} lignes.".format(df.shape[0]))
print("Le dataframe contient maintenant {} colonnes.".format(df.shape[1]))
```

```python
# Compter les lignes du dataframe où une valeur est non renseignée ('NaN')
nans = df.shape[0] - df.dropna().shape[0]
print("%d lignes contiennent un NaN" % nans)
```

```python
# Compter et afficher les lignes où la valeur de 'soutiens' est non renseignée ('NaN')
print("%d lignes contiennent un NaN dans la colonne 'soutiens'." % len(df[df['soutiens'].isnull()]))
df[df['soutiens'].isnull()]
```

```python
# Compter et afficher les lignes où la valeur de 'soutiens' est renseignée
print("%d lignes sont renseignées à la colonne 'soutiens'." % len(df[df['soutiens'].notnull()]))
df[df['soutiens'].notnull()]
```

```python
# Remplir les cellule NaN avec un 0
df.fillna(value = 0, inplace = True)
df
```

```python
# Trier le dataframe par le nombre de 'buts_marqués' (du plus au moins)
df.sort_values('buts_marqués', ascending = False, inplace = True)
df.head()
```

```python
# Réindexer suite à ce tri
df.index = range(1,len(df.index) + 1)
df.head()
```

```python
# Sélectionner les avants ('forward') évoluant à Arsenal

df[ (df['equipe'] == 'arsenal') & (df['poste'] == 'forward') ]
```
